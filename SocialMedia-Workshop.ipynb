**ให้สร้างโมเดลด้วยตัวเองโดยใช้ TF-IDF และ LIWC features ที่มีให้ใน Dataset เพื่อใช้ในการ Train และ Test โมเดลด้วยตัวเอง จากนั้นให้ทำการลด Dimensions ของข้อมูลเพื่อเปรียบเทียบประสิทธิภาพ**

```python
pca = PCA(n_components=40, whiten=True) # n_components สามารถแก้ไขจำนวน Dimensions ตามต้องการได้

X = df.iloc[:, 4:350]
y = df['label']



X = pca.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=42)

log = LogisticRegression(random_state=0).fit(X_train, y_train)

print(log.predict(X_test[:10]))
print(log.predict_proba(X_test[:10]))
print(log.score(X_test, y_test))
```

```python
# ระบุ y_true และ y_pred ด้วยตัวเองให้ถูกต้อง เพื่อใช้แสดงประสิทธิภาพของโมเดล


y_true = y_test
y_pred = log.predict(X_test)



target_names = ['MentalHealth', 'Teaching']
print(classification_report(y_true, y_pred, target_names=target_names))
```

**สร้างกราฟเพื่อศึกษาประสิทธิภาพของโมเดลเมื่อลด Dimensions ของข้อมูล**

```python
dimention = [100,90,80,70,60,50,40,30,20,10]
X = df.iloc[:, 4:350]
y = df['label']
test = []
for n in dimention:
    pca = PCA(n_components=n, whiten=True) # n_components สามารถแก้ไขจำนวน Dimensions ตามต้องการได้

    X = pca.fit_transform(X)
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.30, random_state=42)

    log = LogisticRegression(random_state=0).fit(X_train, y_train)
    score = log.score(X_test, y_test)
    test.append(score)
```

```python
import matplotlib.pyplot as plt
# epochs = range(1, len(test) + 1)

# Define a subplot
# fig, axs = plt(1,figsize=(15,4))
fig = plt.figure()

# Plot loss
# loss_plot = plt.fig

plt.plot(dimention, test, 'c--')
plt.title('Training score with each data dimension')
plt.xlabel('dimension')
plt.ylabel('training score')
plt.legend()
```

จากกราฟด้านบน แสดงประสิทธิภาพของโมเดลเมื่อลด Dimensions ของข้อมูลตั้งแต่ 100 จนถึง 10 จะเห็นว่า โมเดลมี training score ที่ต่ำลง เมื่อทำการลด Dimensions ของข้อมูล แสดงว่า Dimensions ที่มากกว่า จะทำให้โฒเดลมีประสิทธิภาพมากกว่า
